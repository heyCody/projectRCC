```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
from sklearn.cluster import KMeans
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load pre-trained VGG16 model + higher level layers
model = VGG16(weights='imagenet', include_top=False)

# Directory containing unlabelled images (replace with your dataset path)
image_directory = 'D:/sample/archive (1)/sign lang'

# Function to load and preprocess image
def load_and_preprocess_img(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = preprocess_input(img_array)
    return img_array

# Function to extract features using VGG16
def extract_features(img_path):
    img = load_and_preprocess_img(img_path)
    features = model.predict(img)
    features = features.flatten()  # Flatten the features for clustering
    return features

# Extract features for all images in the directory
image_features = []
image_paths = []

for img_file in os.listdir(image_directory):
    img_path = os.path.join(image_directory, img_file)
    image_paths.append(img_path)
    features = extract_features(img_path)
    image_features.append(features)

# Convert the list of image features to a NumPy array
image_features = np.array(image_features)

# Use KMeans to cluster the images based on their features
n_clusters = 5  # Set the number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(image_features)

# Assign each image to a cluster
cluster_labels = kmeans.labels_

# Visualize the clusters using PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(image_features)

# Plot the clusters
plt.figure(figsize=(10, 7))
for i in range(n_clusters):
    cluster_points = pca_result[cluster_labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i}')
    
plt.legend()
plt.title('Image Clusters')
plt.show()

# Optionally: Show images from one cluster
def show_cluster_images(cluster_num, num_images=5):
    cluster_indices = np.where(cluster_labels == cluster_num)[0]
    plt.figure(figsize=(10, 10))
    for i, idx in enumerate(cluster_indices[:num_images]):
        img = image.load_img(image_paths[idx], target_size=(224, 224))
        plt.subplot(1, num_images, i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()

# Display some images from Cluster 0
show_cluster_images(cluster_num=0)


# Manually label clusters based on visual inspection
cluster_labels_to_names = {
    0: 'A',
    1: 'B',
    2: 'C',
    3: 'D',
    4: 'E'
}

# Function to assign cluster labels to the images
def assign_labels_to_images(cluster_labels, cluster_labels_to_names):
    image_to_label = {}
    for i, cluster_num in enumerate(cluster_labels):
        image_to_label[image_paths[i]] = cluster_labels_to_names[cluster_num]
    return image_to_label

# Create a dictionary mapping each image path to its assigned label
image_labels = assign_labels_to_images(cluster_labels, cluster_labels_to_names)

# Print a few examples of image paths and their assigned labels
for img_path, label in list(image_labels.items())[:10]:
    print(f'Image: {img_path} -> Label: {label}')
```
